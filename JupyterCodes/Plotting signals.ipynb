```python
from scipy import signal
```

```python
t = np.linspace(0,1.0,num=2000)
```

```python
len(t)
```

```python
sig_5Hz = np.sin(2*np.pi*5*t)     #  2 * pi * freq=5 * time = t

sig_250Hz = np.sin(2*np.pi*250*t)
```

```python
plt.figure(figsize=(10,10))

plt.plot(sig_5Hz, color='r')
plt.show()
```

```python
plt.figure(figsize=(20,10))
plt.plot(sig_250Hz, color='y')
plt.show()
```

```python
# combining signal

combined_signal = sig_5Hz + sig_250Hz
plt.figure(figsize=(20,10))
plt.plot(combined_signal)
plt.show()
```

```python
style.use('ggplot')
style.use('dark_background')

f, plt_arr = plt.subplots(3,sharex=True)
f.suptitle("Plotting signals")

plt_arr[0].plot(sig_5Hz,color='r',)
plt_arr[0].set_title("5 Hz Sine Wave",color='r')

plt_arr[1].plot(sig_250Hz, color='magenta')
plt_arr[1].set_title("250 Hz Sine Wave", color='magenta')

plt_arr[2].plot(combined_signal, color='b')
plt_arr[2].set_title("Combined Signal")

plt.show()
```
